```python
# Compute the x and y coordinates for points on a sine curve
SepalLengthCm = np.arange(0, 4 * np.pi)
Species = np.sin(SepalLengthCm)
plt.title("sine wave form")

# Plot the points using matplotlib
plt.plot(SepalLengthCm, Species)
```

```python
SepalLengthCm = np.arange(0, 5 * np.pi, 0.1)
y_sin = np.sin(SepalLengthCm)
y_cos = np.cos(SepalLengthCm)

# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
plt.subplot(2, 1, 1)

# Make the first plot
plt.plot(SepalLengthCm, y_sin,'r--')
plt.title('Sine')

# Set the second subplot as active, and make the second plot.
plt.subplot(2, 1, 2)
plt.plot(SepalLengthCm, y_cos,'g--')
plt.title('Cosine')

# Show the figure.
plt.show()
```
